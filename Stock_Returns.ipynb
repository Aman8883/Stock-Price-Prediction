## pct_change returns percentage change of the dataframe
resample allows to change sample frequency. Monthly in this cse


```python
nifty_daily_returns = stk["Adj Close"].pct_change()
nifty_mobthly_returns = stk["Adj Close"].resample("M").ffill().pct_change()
```

```python
plt.plot(nifty_daily_returns)
plt.title("Stock's Daily Returns")
```

## This graph shows that the stock might give a positive return of either 2% or -3% on a daily basis

```python
plt.plot(nifty_mobthly_returns)
plt.title("Stock's MOnthly Returns")
```

## This graph depicts that stock might give a return between -4 to 4 percent monthly

```python
nifty_cumulative_returns = (nifty_daily_returns+1).cumprod()
nifty_cumulative_returns
```

```python
plt.plot(nifty_cumulative_returns)
plt.title("Stock's Cumulative returns")
plt.xlabel("Time")
plt.ylabel("1 Rupee grown into")
```
